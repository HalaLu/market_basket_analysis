```python
import pandas as pd
import numpy as np
from tqdm import tqdm

tqdm.pandas()

import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv('final_data_clusters.csv', index_col=0)
print(data.shape)
data.head()
```

```python
users = list(data['user_id'].unique())
len(users)
```

```python
orders = list(data['order_id'].unique())
len(orders)
```

```python
# data = data.set_index('user_id')
data = data.drop(columns='order_id')
data = data.drop_duplicates(subset=['user_id', 'order_number'])
cols = data.columns
```

```python
data = data.sort_values(['user_id', 'order_number'])
```

```python
print(data.shape)
data.head()
```

```python
indexed = data.set_index(['user_id', 'order_number']).drop(columns=['reorders', 'organics', 'basket_size','order_dow', 'order_hour_of_day', 'days_since_prior_order', 'days_since_first_order', 'num_orders', 'global_reorders', 'tx_purpose_weekly', 'tx_purpose_monthly', 'clusters'])
i_cols = indexed.columns
```

```python
data.columns
```

```python
def get_reorders(row):
    uid = row['user_id'].astype(int)
    onum = row['order_number'].astype(int)
    if onum+1 in indexed.loc[uid].index:
        val = indexed.loc[(uid, onum+1), i_cols]
        return val
    else:
        return [None for  i in i_cols]
```

```python
data[i_cols] = data.progress_apply(get_reorders, axis=1)
```

```python
data = data.dropna()
```

```python
print(data.shape)
data.head()
```

```python
cs = list(set(cols) - set(['user_id', 'clusters', 'Others']))
```

```python
for c in cs:
    l = len(data[c].unique())
    if l > 3:
        print(c, l)
```

```python
c_bins = {}
for c in cs:
    l = len(data[c].unique())
    if l > 5:
        if c == 'order_number':
            data[c], c_bins[c] = pd.qcut(data[c], 10, duplicates='drop', labels=False, retbins=True)
        else:
            data[c], c_bins[c] = pd.qcut(data[c], 5, duplicates='drop', labels=False, retbins=True)

        data[c] = data[c].map({i:f'bin_{i}' for i in range(len(data[c].unique()))})
```

```python
c_bins
```

```python
data['tx_purpose_weekly'] = data['tx_purpose_weekly'].astype(int).map({0:'No', 1:'Yes'})
data['tx_purpose_monthly'] = data['tx_purpose_monthly'].astype(int).map({0:'No', 1:'Yes'})
```

```python
data['clusters'] = data['clusters'].apply(lambda x: f'cluster_{x}')
```

```python
print(data.shape)
data.head()
```

```python
group_cols = ['order_number', 'order_dow', 'order_hour_of_day', 'days_since_prior_order', 'days_since_first_order', 'num_orders', 'global_reorders', 'tx_purpose_weekly', 'tx_purpose_monthly', 'clusters']
drop_cols = ['user_id' ]
```

```python
grouped = data.drop(columns=drop_cols).groupby(group_cols).mean().progress_apply(lambda x: x*100/sum(x), axis=1)
```

```python
print(grouped.shape)
grouped.head()
```
